**Prediction Function**

```python
def predict_price(age,kms,mlg,eng,pwr,sts,car_name,loc,ft,trans,own):
    input_val = np.zeros(xtrain.shape[1])
    
    input_val[0] = age
    input_val[1] = np.log(kms)
    input_val[2] = mlg
    input_val[3] = eng
    input_val[4] = pwr
    input_val[5] = sts
    
    input_val[list(xtrain.columns).index(car_name.lower())] = 1
    input_val[list(xtrain.columns).index(loc)] = 1
    input_val[list(xtrain.columns).index(ft)] = 1
    input_val[list(xtrain.columns).index(trans)] = 1
    input_val[list(xtrain.columns).index(own)] = 1
    
    return np.exp(model1.predict([input_val])) 
```
